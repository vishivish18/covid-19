```python
import pandas as pd
import requests
import io
import datetime
import pymongo
```

```python
base_path_john_hopkins_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'

first_available_data_date = datetime.date(2020,1,22)
today = datetime.date.today()

days_since_first_data = (today-first_available_data_date).days
date_iterator = first_available_data_date
for day in range(days_since_first_data):
    
    date_iterator_month = str(date_iterator.month) if len(str(date_iterator.month)) > 1 else '0'+str(date_iterator.month)
    date_iterator_day = str(date_iterator.day) if len(str(date_iterator.day)) > 1 else '0'+str(date_iterator.day)
    date_iterator_year = str(date_iterator.year)
    valid_file_name = date_iterator_month + '-' + date_iterator_day + '-' + date_iterator_year +'.csv'
    
    print(day," st day valid date is:",valid_file_name)
    r = requests.get(base_path_john_hopkins_url+valid_file_name)
    if r.ok:
        data = r.content.decode('utf8')
        df = pd.read_csv(io.StringIO(data))
        df.to_csv('data/'+valid_file_name,index=False)
    date_iterator = date_iterator + datetime.timedelta(days=1)
        
   


# print(today)

# d = datetime.datetime(2020,1,2)
# d.month
# d.day
# nextday = d + datetime.timedelta(days = 5) 

# print(nextday)

# print((nextday-d).days)
```

### Reading daily data to get India's count

```python
def get_daily_data_for_country(country,df,valid_date, first_case_found):
    try:
        value = df[df['Country/Region']==country]['Confirmed']
        if len(value) > 0:
            print("Total number of cases in Country/Region", country, ' on ', valid_date, ' was ', df[df['Country/Region']==country]['Confirmed'].values[0])
            return (df[df['Country/Region']==country]['Confirmed'].values[0]), valid_date
        else:
            print('No cases found in Country/Region ', country, ' on ', valid_date)
            return 0, valid_date
    except:
        value = df[df['Country_Region']==country]['Confirmed']
        if len(value) > 0:
            print("Total number of cases in Country_Region", country, ' on ', valid_date, ' was ', df[df['Country_Region']==country]['Confirmed'].values[0])
            return (df[df['Country_Region']==country]['Confirmed'].values[0]), valid_date
        else:
            print('No cases found in Country_Region ', country, ' on ', valid_date)
            return 0, valid_date
    
        
    
    
```

```python
days_since_first_data = (today-first_available_data_date).days
date_iterator = first_available_data_date
first_case_found = False
previous_count = 0;

for day in range(days_since_first_data):
    date_iterator_month = str(date_iterator.month) if len(str(date_iterator.month)) > 1 else '0'+str(date_iterator.month)
    date_iterator_day = str(date_iterator.day) if len(str(date_iterator.day)) > 1 else '0'+str(date_iterator.day)
    date_iterator_year = str(date_iterator.year)
    valid_date = date_iterator_month + '-' + date_iterator_day + '-' + date_iterator_year
    valid_file_name = date_iterator_month + '-' + date_iterator_day + '-' + date_iterator_year +'.csv'
    
    #print(day," st day valid date is:",valid_file_name)
    df = pd.read_csv('data/'+valid_file_name)
    country = 'India'
    total_cases, date_of_incident = get_daily_data_for_country(country,df,valid_date,first_case_found)
    print('delta is', (total_cases-previous_count))
    previous_count = total_cases
    
    if total_cases > 0:
        if first_case_found == False:
            print('Frist case for', country,  'on' ,date_of_incident )
            first_case_found = True
    
        
    date_iterator = date_iterator + datetime.timedelta(days=1)

```
